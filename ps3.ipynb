```python
import pandas as pd
import numpy as np
import datetime as dt            #importing necessary libraries

url="https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data"        #importing data from url
data=pd.read_csv(url,sep='\s+')      
data.head(10)               #displaying first 10 rows because it is a big data file
```

```python
data["date"]=pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('/'.join,axis=1))     #creating new column and assigning values of first three columns as date
data=data.drop(columns=['Yr','Mo','Dy'])       #deleting first three columns
data.head(10)      #displaying the dataframe
```

```python
data["date"]=np.where(pd.DatetimeIndex(data["date"]).year<2000,data.date,data.date-pd.offsets.DateOffset(years=100))
data.head(10)             #changing date from 2000s to 1900s
```

```python
data2=data.set_index("date")
data2.index.astype("datetime64[ns]")         #resetting data type as datetime64[ns]
```

```python
null=data2.isnull().sum()        #creating array of all null values
null.count()                     #counting the total null values
```

```python
non_missing=data2.count()        #calculating total non missing values
non_missing.sum()                #displaying total non missing values
```

```python
wind=data2.mean()            #calculating mean windspeed
wind.mean()                  #displaying the mean windspeed
```

```python
def values(sample):                 #defining function
    sample=pd.Series(sample)        
    Min=sample.min()                #calculating min
    Max=sample.max()                #calculating max
    Mean=sample.mean()              #calculating mean
    Sd=sample.std()                 #calculating standard deviation
    result=[Min,Max,Mean,Sd]        #storing the result in array
    ind=["Min","Max","Mean","Sd"]   #giving index
    result=pd.Series(result,index=ind)
    return result
loc_stats=data2.apply(values)                #creating dataframe loc_stats
loc_stats                                    #displaying the dataframe
```

```python
day_stats=data2.apply(values,axis=1)      #creating dataframe to calculate min max mean and SD of each day 
day_stats.head(10)         #displaying first 10 rows of dataframe
```

```python
january=data2[data2.index.month==1]            #Verage windspeed of january 
january.mean()
```

```python
data2.resample('Y').mean()          #resample data yearly
```

```python
data2.resample('M').mean()       #resample data monthly
```

```python
data2.resample('W').mean()       #resample data weekly
```

```python
year_1961=data2[data2.index.year==1961]                #setting year index to 1961
stats=data2.resample('W').mean().apply(lambda sample:sample.describe())  #getting weekly data and other functional data from the previously described sample function
stats
```

```python
import numpy as np
import matplotlib.pyplot as plt

address="us-marriages-divorces-1867-2014.csv"
data=pd.read_csv(address)

year1=data[data['Year']==1900]
year2=data[data['Year']==1950]
year3=data[data['Year']==2000]
all_years=[year1,year2,year3]

new_data=pd.concat(all_years)
new_data.plot(x="Year",y=['Marriages','Divorces'],kind="bar")
plt.xlabel('Year')
plt.ylabel('Number')
plt.title('Number of marriages and divorces in US per capita between 1900, 1950 and 2000')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

address="roman-emperor-reigns.csv"
data=pd.read_csv(address)

asntd=data.Cause_of_Death.where(data.Cause_of_Death=='Assassinated').count()
total=data.Cause_of_Death.count()
asntd_frac=(asntd/total)*100
others_frac=100-asntd
Data=[asntd_frac,others_frac]
label=('Assassinated','Others')
plt.pie(Data,labels=label,autopct='%.2f%%')
plt.title('Fraction of all Roman Emperors that were assassinated')
plt.show()
```
